```python
from csv import reader

aktoriai = []

with open('./Sakila_actor_films.csv') as failas:

    csv_reader = reader(failas)
    next(csv_reader)
    for eilute in csv_reader:
        aktorius = {
            'first_name' : eilute[1],
            'last_name' : eilute[2],
            'title' : eilute[4],
            'rental_rate' : float(eilute[5]),
            'replacement_cost' : float(eilute[7]),
            'rating' : eilute[8],
            'category_name' : eilute[9],
        }
        aktoriai.append(aktorius)

print(aktoriai)
```

```python
#Didžiausia nuomos kaina:

max_rental_rate = aktoriai[0]

for rental_rate in aktoriai:
    if rental_rate['rental_rate'] > max_rental_rate['rental_rate']:
        max_rental_rate = rental_rate

print('Didžiausia nuomos kaina yra', max_rental_rate['rental_rate'])
```

```python
#Nuomos kainos vidurkis:

avg_rental_rate = [rental_rate['rental_rate'] for rental_rate in aktoriai]

suma = sum(avg_rental_rate)
kiekis = len(avg_rental_rate)
vidurkis = round(suma / kiekis)

print('Nuomos mokesčio vidurkis:', vidurkis)
```

```python
#Mažiausia nuomos kaina:

min_rental_rate = aktoriai[0]

for rental_rate in aktoriai:
    if rental_rate['rental_rate'] < min_rental_rate['rental_rate']:
        min_rental_rate = rental_rate

print('Mažiausia nuomos kaina yra', min_rental_rate['rental_rate'])
```

```python
#Pilna nuomos kainos suma:

rental_rate_sum = [rental['rental_rate'] for rental in aktoriai]
suma = sum(rental_rate_sum, 2)

print('Nuomos kainos pilna suma', round(suma, 2))
```

```python
#Nuomos kainos grafikas:

x = np.array(['Maž. nuomos kaina', 'Nuomos kainos vid.', 'Did. nuomos kaina'])
y = np.array([0.99, 3.0, 4.99])

plt.title('Nuomos mokesčio grafikas')
plt.bar(x,y, color = "#4CAF90", width = 0.9)
plt.show()
```

```python
#Eilučių ir žodžių kiekis:

eilutes = []

with open('./Sakila_actor_films.csv') as failas:
    for eilute in failas:
        eilute = eilute.rstrip('\n')
        eilutes.append(eilute)

print('Eiluciu kiekis:', len(eilutes))

kiek_zodziu = sum([ len(eilute.split()) for eilute in eilutes ])
print('Iš viso žodžių:', kiek_zodziu)
```
